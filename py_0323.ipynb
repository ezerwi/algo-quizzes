# BOJ / 2839. 설탕 배달

```
메모리      시간        코드길이

29284kb     56ms	138B
```

초딩 때 문제 같은데 수학적으로 생각하기가 ..

````
n = int(input())
ans = -1
for i in range((n//5), -1, -1) :
    if (n-5*i)%3 == 0 : 
        ans = i + (n-5*i)//3
        break
print (ans)
````
