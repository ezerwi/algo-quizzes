# Silver IV. 소수 찾기

https://www.acmicpc.net/problem/1978

```
N = int(input())
nums =map(int, input().split())

def isPrime(num:int) :
    n = 2
    while n*n <= num:
        if (num % n == 0) : return False
        n += 1
    return True

ans = 0

for n in nums:

    if n == 1 : continue
    elif n == 2 or isPrime(n): 
        ans += 1

print (ans)
```

## Short Codings

https://www.acmicpc.net/source/6548721
### `all(iterable)`

iterable 내부 elements 가 모두 True 인 경우 또는 빈 객체인 경우 return True

```
input()
print(
    len([p for p in map(int, input().split()) 
        if p != 1 
            and all(p % i for i in range(2, p))])
    )
```
