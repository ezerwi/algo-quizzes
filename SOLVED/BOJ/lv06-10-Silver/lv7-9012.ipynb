# Silver4. 괄호

https://www.acmicpc.net/problem/9012

```python
def isVPS(line) :
    stack = []
    for l in line:
        if l == '(': stack.append(l)
        else:
            if not stack : 
                return 'NO'
            stack.pop()
    if stack: return 'NO'
    return 'YES'            

for _ in range(int(input())):
    line = list(input())    
    print (isVPS(line))
```

## Short Codings

https://www.acmicpc.net/source/12321280

짧고 직관적 코드, 성능은 크게 뒤쳐지지 않음

```python
for _ in'_'*int(input()):
    s=input()
    for _ in s:s=s.replace('()','')
    print(['YES','NO'][s>''])
```
