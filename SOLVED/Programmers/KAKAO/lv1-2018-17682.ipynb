```
import re 

def getThree(result: str) -> [str]:
    return re.findall('\d+[SDT][\*\#]{0,1}', result)


def getOneScore(a: str) -> int:
    num = int(re.search('\d+', a).group())
    SDT = re.search('[SDT]', a).group()
    starOrMinus = re.search('[\*\#]', a)
    option = starOrMinus.group() if starOrMinus else False
    
    return [num ** (1 if SDT == 'S' else 2 if SDT == 'D' else 3),  2 if option == '*' else -1 if option == '#' else 1]


def solution(dartResult: str) -> int :
    a, b, c = getThree(dartResult)
    scoreA, optionA = getOneScore(a)
    scoreB, optionB = getOneScore(b)
    scoreC, optionC = getOneScore(c)
    options = [optionA, optionB, optionC]
    
    for idx, op in enumerate(options):
        if op == 2 and idx > 0:
            options[idx-1] *= op

    return scoreA * options[0] + scoreB * options[1] + scoreC * options[2]


qq = [
    '1S2D*3T',
    '1D2S#10S',
    '1D2S0T',
    '1S*2T*3S',
    '1D#2S*3S',
    '1T2D3D#',
    '1D2S3T*',
]


for q in qq:
    print (solution(q))
```
