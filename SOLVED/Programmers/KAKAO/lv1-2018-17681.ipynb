```
def getBinary(num: int, size: int) -> str:
    return f"{num:{size}b}".replace(' ', '0')


def getBinaryArr(arr: [str], size: int) -> [str]:
    binaryArr = []
    for num in arr:
        binaryArr.append(getBinary(num, size))
    return binaryArr


def concatTwoBinary(bin1: str, bin2: str) -> str :
    concat = tuple(zip(bin1, bin2))
    result = ''
    for a, b in concat:
        result += '#' if a == '1' or b == '1' else ' '
    return result
        

def solution(n:int, arr1: [int], arr2: [int]) -> [str]:
    concatArray = tuple(zip(getBinaryArr(arr1, n), getBinaryArr(arr2, n)))
    result = []
    for a, b in concatArray:
        result.append(concatTwoBinary(a, b))
    return result


qq = [
    [5, [9, 20, 28, 18, 11], [30, 1, 21, 17, 28]],
    [6, [46, 33, 33 ,22, 31, 50], [27 ,56, 19, 14, 14, 10]]
]

for q in qq:
    [n, arr1, arr2] = q
    print (solution(n, arr1, arr2))
```
