풀긴했는데 뭔가 좀 지저분하다

```
qs = [
    [3, [[1, 1, 0], [1, 1, 0], [0, 0, 1]]],
    [3, [[1, 1, 0], [1, 1, 1], [0, 1, 1]]]
]
```

```
def dfs (graph, v, visited, network) :
    visited[v] = True
    network.append(v)
    for idx, val in enumerate (graph[v]):
        if idx == v : continue
        if not visited[idx] and val == 1 :
            dfs(graph, idx, visited, network)
    return network

def solution(n, computers):

    networks = []    
    visited = [False] * n
    
    for c in range(n):
        networks.append (dfs(computers, c, visited, []))

    ans = []
    sortedByLen = sorted(networks, key=lambda x: len(x))
    
    while sortedByLen:
        curr = sortedByLen.pop()
        ans.append(curr)
        for c in curr:
            for net in sortedByLen:
                if c in net: net.remove(c) 

    return len ([a for a in ans if len(a) > 0])


for q in qs:
    print (solution(q[0], q[1]))
```

아래 풀이가 가장 정석인 듯

```
def solution(n, computers):
    answer = 0
    visited = [0 for i in range(n)]

    def dfs(computers, visited, start):
        stack = [start]
        
        while stack:
            j = stack.pop()
            if visited[j] == 0:
                visited[j] = 1

            for i in range(0, len(computers)):
                if computers[j][i] ==1 and visited[i] == 0:
                    stack.append(i)

    i=0
    while 0 in visited:
        if visited[i] ==0:
            dfs(computers, visited, i)
            answer +=1
        i+=1
    return answer
```
