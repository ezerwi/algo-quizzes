# BOJ/1065. 한수

input 이 1000 이하인 것을 놓치지 않으면 좀더 빨리 풀 수 있었을 거 같다ㅜ

메모리  시간	코드 길이

29284	52		222

```
count = 0

for i in range(1, int (input())+1) :
    num = [ int(x) for x in str(i) ]
    
    if len(num) < 3 : 
        count += 1
        continue
    
    if num[2] - num[1] == num[1] - num[0] : 
        count += 1
        # print (num[s], num[s-1], num[s-2])
print (count)
```
