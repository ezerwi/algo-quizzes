# BOJ/1065. 한수

input 이 1000 이하인 것을 놓치지 않으면 좀더 빨리 풀 수 있었을 거 같다ㅜ

메모리  시간	코드 길이

29284	52		222

```
count = 0

for i in range(1, int (input())+1) :
    num = [ int(x) for x in str(i) ]
    
    if len(num) < 3 : 
        count += 1
        continue
    
    if num[2] - num[1] == num[1] - num[0] : 
        count += 1
        # print (num[s], num[s-1], num[s-2])
print (count)
```

# BOJ/1157. 단어 공부

지저분하게 푸럿다..

```
word = [i for i in str(input())]
temp = [0 for i in range(26)]
for _ in word :
    temp[ord(_.upper())-ord('A')]+=1
res = temp.index(max(temp))+ord('A')
if temp.count(max(temp)) >1 :
    print('?')
else : print (chr(res))
```

## Short Coding

기준과 비교대상을 바꿔서 훨씬 깔끔하게 풀이..

풀이가 지저분해진다 싶으면 기준을 바꿔봐야겠다.


```
s,a=input().lower(),[]
for i in range(97,123):
 a.append(s.count(chr(i)))
print('?'if a.count(max(a))>1 else chr(a.index(max(a))+97).upper())
```

# BOJ/Silver V/1316. 그룹 단어 체커
메모리	시간	코드길이

29284kb	56ms	269B


1차 시도에서 문제를 제대로 이해 못하고 엉뚱하게 풀어서 바로 틀림..


```
count = 0

for _ in range(int(input())) :
    cur = [i for i in input()]
    cri = len(set(cur))
    temp = 0
    for x in range(len(cur)-1, 0, -1) :
        if cur[x-1] != cur [x] : 
            temp +=1
    if temp == cri-1 :
        count += 1
        
print (count)
```

# Goorm/3-2주차/lv2. 경품 추첨

다른 언어였으면 좀 힘들었을듯..

python에서는 dictionary가 있기 때문에, 비교가 훨씬 쉬운 듯

```
nm = list(map(int, input().split()))
n = {}

for i in list(map(str, input().split())) :
	if i not in n : n[i] = 1
	else : n[i] += 1
# print (n)
for j in list(map(str, input().split())) :
	print (n.get(j) if j in n else 0, end=" ")
```
