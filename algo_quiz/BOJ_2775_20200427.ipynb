# [BOJ > 2775. 부녀회장이 될테야 ](https://www.acmicpc.net/problem/2775)

어설픈 재귀..

시간초과!

```python
def room(n, k):
    if n == 0 : return k
    if k == 0 : return 0
    elif k == 1: return 1
    # if (n, k) in rmlist : return rmlist
    print (n, k)
    return room(n, k-1) + room (n-1, k)

for i in range(int(input())) :
    k, n = int(input()), int(input())
    # print (k)
    # print (n)

    print (room (k, n))
```

이 간단한거 하는데 시간이 왜케 걸렸지..ㅋㅋ

```python
res = [[0 for i in range(15)] for j in range(15)]
res[0] = [i for i in range(15)]
for i in range(1, 15) :
    for j in range(1, 15) :
        res[i][j] = res[i-1][j] + res[i][j-1]
        # print (i, j)

display (res)
    
for i in range(int(input())) :
    k, n = int(input()), int(input())
    # print (k)
    # print (n)

    print (res [k][n])
```
